```python
import math

import pandas as pd

# Prepare data, exclude Rumah X
data = {
    "rumah": ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"],
    "lat": [11, 15, 19, 18, 16, 23, 25, 21, 23, 29],
    "long": [26, 29, 28, 30, 26, 25, 22, 24, 25, 24],
    "lokasi": [
        "kota",
        "kota",
        "kota",
        "kota",
        "kota",
        "kabupaten",
        "kabupaten",
        "kabupaten",
        "kabupaten",
        "kabupaten",
    ],
}

# Buat DataFrame
df = pd.DataFrame(data)

# Definisikan titik target
target = [19, 25]


# Fungsi untuk menghitung jarak Euclidean
def calculate_distance(row):
    return math.sqrt((row["lat"] - target[0]) ** 2 + (row["long"] - target[1]) ** 2)


# Tambahkan kolom baru dengan jarak
df["jarak"] = df.apply(calculate_distance, axis=1)

df
```

```python
# Sort by lat and then by long
sorted_df = df.sort_values(by=["jarak"])

# Tampilkan DataFrame yang diperbarui
sorted_df
```
